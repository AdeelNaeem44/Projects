```python
reteriver=vectorstore.as_retriever(search_type="similarity",search_kwargs={"k": 4})
```

```python
from google.generativeai import GenerativeModel
llm = genai.GenerativeModel(model_name="models/gemini-1.5-flash-latest")

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="you are a helpful assistant answer only the question based on the context provided. If you don't know the answer, say 'I don't know'.\n\nContext: {context}\n\nQuestion: {question}\n\nAnswer:",
)
```

```python
question='is the topic of nuclear fusion were discussed in this video? If yes, then what is the main point of discussion?'
reteriver_docs=reteriver.invoke(question)
```

```python
print(reteriver_docs[0].page_content)
```

```python
context = "".join([doc.page_content for doc in reteriver_docs])
final_prompt = prompt.invoke({"context": context, "question": question})
prompt_str = final_prompt.to_string()
print(prompt_str)
```

```python
response = llm.generate_content(prompt_str)
answer = response.candidates[0].content.parts[0].text
print(f"Answer: {answer}")  
```

```python
from langchain_core.runnables import RunnablePassthrough , RunnableLambda ,RunnableParallel
from langchain_core.output_parsers import StrOutputParser

def format_docs(reteriver_docs):
    context_text= " ".join(doc.page_content for doc in reteriver_docs)
    return context_text
```

```python
parallel_chain=RunnableParallel(
    {
        "context":reteriver| RunnableLambda(format_docs),
        "question":RunnablePassthrough()
    }
)
```

```python
parallel_chain.invoke("who is Demis")
```

```python
from langchain_core.runnables import RunnableLambda

def call_gemini(input_dict):
    prompt_text = f"""
You are a helpful assistant. Answer the question based only on the context provided.

Context:
{input_dict['context']}

Question:
{input_dict['question']}

Answer:"""
    response = llm.generate_content(prompt_text)
    return response.candidates[0].content.parts[0].text.strip()

gemini_runnable = RunnableLambda(call_gemini)  # ✅ Correct use
```

```python
main_chain = parallel_chain | gemini_runnable

```

```python
def get_response_from_chain(query: str) -> str:
    result = main_chain.invoke(query)
    return result

```
